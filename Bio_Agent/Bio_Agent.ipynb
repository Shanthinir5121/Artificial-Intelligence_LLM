```python
import os
import getpass
from openai import OpenAI
```

```python
# Enter your API key
os.environ["OPENAI_API_KEY"] = getpass.getpass("Enter your OpenAI API key: ")
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])
```

```python
def extract_with_llm(question, user_input):
    prompt = f"""
You are an information extractor.
Extract the exact and clean answer to the question below.

QUESTION: {question}
USER RESPONSE: {user_input}

RULES:
- Remove filler words like "my name is", "it is", "I live in".
- For email → return only valid email address.
- For phone number → return only 10-digit phone number.
- For address → return only place + pincode (if given).
- DO NOT add extra words. Only return the extracted answer.

Return ONLY the clean answer.
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content.strip()
```

```python
def get_personal_details():

    print("\nHi, help me with your full name")
    raw_name = input("")
    name = extract_with_llm("Extract user's full name.", raw_name)

    print("\nWhat is your date of birth?")
    raw_dob = input("")
    dob = extract_with_llm("Extract user's date of birth.", raw_dob)

    print("\nWhat is your email address?")
    raw_email = input("")
    email = extract_with_llm("Extract user's email address.", raw_email)

    print("\nWhat is your phone number?")
    raw_phone = input("")
    phone = extract_with_llm("Extract user's phone number.", raw_phone)

    print("\nWhat is your address?")
    raw_address = input("")
    address = extract_with_llm("Extract user's address (place + pincode only).", raw_address)

    # Initial Display
    print("\n----------------------------------------------------")
    print(f"Name:   {name}")
    print(f"Date of Birth:  {dob}")
    print(f"Email:  {email}")
    print(f"Phone:  {phone}")
    print(f"Address: {address}")
    print("----------------------------------------------------")

    # EDIT LOOP
    while True:
        edit = input("Do you want to change your personal details? (y/n)\n").lower()

        if edit == "n":
            break

        print("Choose edit section: 1. Name, 2. Date of Birth, 3. Email, 4. Phone, 5. Address. Type 'q' to quit")
        section = input("").lower()

        if section == "q":
            break

        if section == "1":
            val = input("Edit Name: ")
            name = extract_with_llm("Extract user's full name.", val)

        elif section == "2":
            val = input("Edit Date of Birth: ")
            dob = extract_with_llm("Extract user's date of birth.", val)

        elif section == "3":
            val = input("Edit Email: ")
            email = extract_with_llm("Extract user's email address.", val)

        elif section == "4":
            val = input("Edit Phone: ")
            phone = extract_with_llm("Extract user's phone number.", val)

        elif section == "5":
            val = input("Edit Address: ")
            address = extract_with_llm("Extract user's address (place + pincode only).", val)

    # FINAL OUTPUT
    print("\n----------------------------------------------------")
    print("Edited Information")
    print("----------------------------------------------------")
    print(f"Name:   {name}")
    print(f"Date of Birth:  {dob}")
    print(f"Email:  {email}")
    print(f"Phone:  {phone}")
    print(f"Address: {address}")
    print("----------------------------------------------------")
```

```python
get_personal_details()
```
