```python
import os
import getpass
from openai import OpenAI
```

```python
os.environ["OPENAI_API_KEY"] = getpass.getpass("Enter your OpenAI API key: ")
client = OpenAI()
```

```python
def process_experience(job, company, start_date, end_date, description):
    prompt = f"""
You are an expert resume strategist with deep experience in crafting
impact-driven, ATS-optimized professional summaries.

Your task is to transform the user’s raw work experience into:

1. High-impact, achievement-focused bullet points written using
   strong action verbs, measurable results, and concise professional language.

2. A refined “Improvement Section” with exactly 3 bullets highlighting:
   – skills strengthened,
   – process improvements,
   – measurable outcomes or growth demonstrated during the role.

Use the information exactly as provided by the user:

INPUT:
Job Title: {job}
Company: {company}
Duration: {start_date} – {end_date}
Description: {description}

Now produce a polished, resume-ready output in the EXACT following format:

Job Title
Company
Start Month Year – End Month Year

- Bullet 1
- Bullet 2
- Bullet 3
- Bullet 4

Improvement Section:
- Improvement 1
- Improvement 2
- Improvement 3
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content
```

```python
all_experiences = []

print("  Work Experience Collector")

while True:
    print("\nEnter a new experience (or type 'quit' at Job Title to exit):")

    job = input("Job Title: ")
    if job.lower().strip() == "quit":
        break

    company = input("Company: ")
    start_date = input("Start Month & Year (e.g., Jan 2022): ")
    end_date = input("End Month & Year (e.g., Dec 2023 or Present): ")
    description = input("Describe your work (2–5 lines): ")

    print("\nProcessing with LLM...")
    final_text = process_experience(job, company, start_date, end_date, description)

    all_experiences.append(final_text)

    print("\nExperience Added Successfully!")
    print("\n------------------------------------------")
    print(final_text)
    print("------------------------------------------")
```

```python
print("\n\n====================================")
print("FULL EXPERIENCE SECTION")
print("====================================\n")

for i, exp in enumerate(all_experiences, 1):
    print(f"Experience {i}:\n{exp}\n")
```
